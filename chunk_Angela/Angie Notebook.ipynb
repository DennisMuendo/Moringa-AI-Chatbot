# Creating A Self-Learning Moringa School AI Chat-Bot 


#  Overview

The aim of this project is to create an AI chatbot for Moringa School's website. The aim of the chatbot is to understand user queries and connect them to key information including courses offered and  enrollment information, and also offer technical support for any issues faced. 

# Business Understanding


A chatbot is a program whose aim is to simulate conversation with human users. [Research](https://www.juniperresearch.com/research/telco-connectivity/messaging/chatbots-trends-research-report/) suggests that more and more companies are adopting the use of chatbots, further estimating that by the end of 2023, more than 75% of customer queries will be resolved by chatbots.


This increase in adoption of chatbot usage can be attributed to their ability to : boost efficiency by automating redundant/ recurring queries,  offer quick(er) customer assistance and improve customer experience by reducing wait time and offering 24/7 support.

In the [education sector](https://onlinelibrary.wiley.com/doi/abs/10.1002/cae.22326), chatbots have been clustered into two categories: "service-oriented chatbots" and "teaching oriented chatbots". Teaching-oriented chatbots provide assistance in academic activities.

Service oreinted chatbots, which is what this project aims to create, are structured to answer Frequently Asked Questions (FAQs) and provide general information. An example can be see from Durham Univeristy's virtual assistant [Holly](https://www.durham.ac.uk/study/ask-us/) and Georgia State University's virtual assistant [Pounce](https://www.durham.ac.uk/study/ask-us/). Pounce in particular performed well, especially in streamlining the admission process.

# Business Problem

Chatbot utilization is not a new concept in Kenya.[Companies](https://axis.africa/articles/3-kenyan-companies-using-whatsapp-chatbots-to-automate-messaging/) in Kenya are using chatbots to answer customer queries and provide access to their services.[Moringa School](https://moringaschool.com/) Nairobi is a learning accelerator that aims to provide digital and professional skills to young Africans. Some of the courses offered include Software Engineering, Data Science and Product Design.

Since its inception in 2014, Moringa has trained over 2000 students,in Kenya and globally, and is set to train [over 200,000 students by 2030](https://solve.mit.edu/challenges/good-jobs-and-inclusive-entrepreneurship/solutions/21696#:~:text=Moringa%20School%20has%20demonstrated%20considerable,directly%20serving%20around%202%2C800%20students). Furthermore, Moringa School has just launched a [new course](https://moringaschool.com/corporate/mpesa-mini-apps-api-development/) and has several new [partnerships](https://moringaschool.com/blog/edtech-leader-moringa-school-signs-partnership-with-nairobi-county-government-to-drive-its-advancement-as-a-global-digital-hub/), all with the aim of equipping young profesisonals with tech skills.

Currently, you can only access Moringa services through visiting the website or calling the relevant departments. As the schools continues to grow, so will the demand for its services. An efficient way to fill this gap would be to employ the use of a chatbot.

The objectives of the chatbot would be:
- Provide fast, 24/7 Service. During peak hours, peak seasons or with international students who are on a different timezone, the chatbot will be able to offer ready support, reducing wait times when human assistants are offline. 


- Improve Customer Experience. The chabtot will be able welcome users to the website and offer navigation assistance by giving links to specific resources. This will make a user's experience easier and efficient. 


- Provide Aceess to Information. Users can ask the chat-bot general questions about Moringa School, such as information about courses offered, admission procedures and tuition fees. The chat-bot can also provide details, schedules, and registration information for events. 


- Offer Technical Support. The chatbot will offer technical support for any issues users might face on the website. It will be able to guide users through troubleshooting steps or redirect them to relevant resources. 


-  Data Collection. As the chatbot provides assistance, it will be able to collect data on how users ask questions, the most frequently asked questions and areas on which it can improve. This data will enable the chatbot to learn and give more accurate and personalized responses. The chatbot will also improve on detecting which queries it can handle and which it can pass for human assistance. 

# Data Understanding

The data used in this project has been obtained by scraping the Moringa School's Websites.
.The data extraction process involved two main Python scripts organized into separate subfolders within the **moringa scraper**  directory.

1. Link Scraper:

**link_scraper.py** contains a fucntion designed to go through the Moringa School website and extract hyperlinks present on its pages.
It utlises the BeautifulSoup library. It navigates through the HTML structure, identifies anchor tags (" a ") containing hyperlinks, and employs the urljoin function to handle URL concatenation. The extracted links are then saved to a JSON file named **scraped_links.json**.
    
2. Web Scraper:

**web_scraper.py** contains a function that utilizes the links obtained from the previous step (scraped_links.json) to scrape text content from the corresponding web pages.
After reading the previously saved links from the JSON file, the script makes HTTP requests to each link, retrieves the HTML content, and uses BeautifulSoup to extract text from specified tags. The extracted text is then saved in a structured format in the **moringa_text_corpus.json file**.
Data Files:

3. scraped_links.json
    
**scraped_links.json** contains a list of unique URLs obtained during the link scraping process.
    
    
4. moringa_text_corpus.json
    
**moringa_text_corpus.json** stores the text content scraped from the corresponding URLs, organized in a dictionary structure where each link is associated with a list of unique text snippets.




# Data Preparation

The goal of this step is to create an **intent list** that is a list of dictionnaries.**Intents** are what the customer intends to do.  An intent list helps the chatbot understand the user's goals. These intents are what are used to train our chatbot, allowing it to understand the different types of user queries and learn how to respond to them effectively. 


We have manually gone through the corpus(moringa text corpus.json) to come up with intents that are unique to Moringa School Website's users. This approach has ensured that there is diversity in the intent list; ie, the queries cover a diverse range of topics(courses offered, admission requirements,fees, etc). Diversity of queries will help the chatbot to be able to handle a wide array of uer queries. 


These initial intent lists are stored in the directories (chunk..+ Angela, Dennnis, Frederick, Kelvin, Mark, Peninah, Samuel). The final intent list is combined from all the different intent lists and stored in in the file **final intents.json**. 

The final intents list has the following structure:
- "tag": The label/ category of the intent.For example, "software_engineering_course_information"
and "cybersecurity_admission" are tags that represent distinct areas of information related to Moringa School. 


- "questions": A list of questions related to the intent representing how users might inquire about something. For example,"What topics does the software enginnering course cover" is a question under the tag "software_engineering_course_information".


- "responses": A list of responses corresponding to the questions.For example, a response to the above question is, "The software engineering course covers HTML and CSS".
